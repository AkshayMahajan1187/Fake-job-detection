```python
missing_percentages_frauds={}
missing_percentage_Nonfrauds={}
for col in columns:
    missing_percentages_frauds[col]=((((df['fraudulent']==1) & df[col].isnull()).sum())/fraudCount)*100
    missing_percentage_Nonfrauds[col]=((((df['fraudulent']==0) & df[col].isnull()).sum())/(df.shape[0]-fraudCount))*100
print(missing_percentages_frauds)
print(missing_percentage_Nonfrauds)
```

```python
compare_missing = pd.DataFrame({
    'fraud_missing_%': pd.Series(missing_percentages_frauds),
    'nonfraud_missing_%': pd.Series(missing_percentage_Nonfrauds)
})

compare_missing['difference_in_missing_percentage'] = (
    compare_missing['fraud_missing_%'] -
    compare_missing['nonfraud_missing_%']
)

compare_missing
```

```python
urgency_keywords = [
    'immediate', 'urgent', 'asap',
    'quickly', 'fast', 'fast hiring', 'limited'
]
for i in range(len(df)):
    text = df.loc[i, 'description']
    
    if pd.isna(text):
        continue  
    
    text = text.lower()
    
    for word in urgency_keywords:
        if word in text:
            df.loc[i, 'has_urgency_language'] = 1
            break

```

```python
count_urgency_fake=((df['fraudulent']==1) & (df['has_urgency_language']==1)).sum()
count_urgency_fake
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
df['org_count'] = 0
df['gpe_count'] = 0

for i in range(len(df)):
    text = df.loc[i, 'company_profile']
    
    if pd.isna(text):
        continue
    
    doc = nlp(text)
    
    for ent in doc.ents:
        if ent.label_ == 'ORG':
            df.loc[i, 'org_count'] += 1
        elif ent.label_ == 'GPE':
            df.loc[i, 'gpe_count'] += 1

df.groupby('fraudulent')[['org_count', 'gpe_count']].mean()
```

```python
df['lexical_diversity']=0.0
for i in range(len(df)):
    text=df.loc[i,'description']
    if pd.isna(text):
        continue
    words=text.lower().split()
    if(len(words)==0):
        continue
    else:
        unique_words=set(words)
        df.loc[i,'lexical_diversity']=len(unique_words)/len(words)
    
```

```python
df.groupby('fraudulent')['lexical_diversity'].mean()
```

```python
df['num_in_requirements']=0.0
for i in range(len(df)):
    text=df.loc[i,'requirements']
    if pd.isna(text):
        continue
    count=0
    for ch in text:
        if ch.isdigit():
            count+=1
        df.loc[i,'num_in_requirements']=count

df.groupby('fraudulent')['num_in_requirements'].mean()
```

EDA Summary & Feature Decisions

- The dataset is highly imbalanced, with only 866 fraudulent postings out of 17,880 (~4.8%). This implies that accuracy alone would be misleading, since a naive model predicting all jobs as real would already achieve ~95% accuracy. Therefore, metrics such as recall and precision will be more meaningful.

- Several features contain a large number of missing values. Missing-value analysis revealed that company_profile shows a very large missingness gap (>50%) between fake and real jobs, making it a strong fraud signal. Other fields such as required_experience and employment_type show smaller but noticeable differences and are treated as secondary signals.

- The length of company_profile text differs substantially between fake and real jobs, with real postings containing significantly longer and more informative descriptions. This suggests that fake postings often avoid providing detailed company information even when the field is present.

- A binary feature has_urgency_language was introduced to capture urgency-related phrasing in job descriptions. While urgency language appears more frequently in fraudulent jobs (~25%) than in real jobs (~20%), the overlap indicates that this feature alone is weak and should only be used in combination with stronger signals.

- Two Named Entity Recognition–based features, org_count and gpe_count, were added to quantify how often organizations and locations are mentioned. Real job postings contain significantly more named entities on average than fake postings, indicating intentional avoidance of verifiable information by fraudulent listings. These features emerged as some of the strongest and most reliable indicators of fraud.

- A lexical diversity feature was explored to measure repetitiveness in job descriptions. However, the observed difference between fake and real jobs was small, suggesting that lexical diversity alone does not strongly distinguish fraudulent postings and is therefore deprioritized.

Final Feature Decision:
The strongest signals identified during EDA are company information availability and specificity, particularly missing company profiles and low counts of named organizations and locations. Supporting features include missing experience requirements and urgency language, while purely stylistic features such as lexical diversity are considered weak.
